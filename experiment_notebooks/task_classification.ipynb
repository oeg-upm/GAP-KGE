```python
import json
import Levenshtein
```

```python
with open("../data/papers_data.json", "r", encoding="utf-8") as f:
    paper_list = json.load(f)

tasks = []

for i, paper in enumerate(paper_list):
    tasks.append(paper['Tasks'])

print(tasks)
```

```python
from collections import Counter

task_list = []

def flatten(list_element):
    if isinstance(list_element, list):
        for task in list_element:
            flatten(task)
    else:
        task_list.append(list_element)

flatten(tasks)

task_frecuency = Counter(task_list)

task_list = [task for task, count in task_frecuency.items() if count > 1]

print("Frecuencias completas:", task_frecuency)
print("Lista actualizada (solo repetidas):", task_list)
```

```python
task_list = [x.lower() for x in task_list]
print(task_list)
```

```python
task_list
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

corpus_embeddings = model.encode(task_list)
clusters = util.community_detection(corpus_embeddings, min_community_size=1, threshold=0.65)

for i, cluster in enumerate(clusters):
    print(f"\nCluster {i+1}:")
    for sentence_id in cluster:
        print(f"  - {task_list[sentence_id]}")
```

```python
target_word = "knowledge"
task_list = [
    " ".join([w for w in s.split() if w.lower() != target_word.lower()]) 
    for s in task_list
]
```

```python
from sentence_transformers import SentenceTransformer, util
import time

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(task_list, convert_to_tensor=True)

start_time = time.time()
clusters = util.community_detection(embeddings, min_community_size=1, threshold=0.62)

print(f"Agrupación completada en {time.time() - start_time:.4f} segundos\n")

for i, cluster in enumerate(clusters):
    print(f"Grupo {i+1}:")
    for index in cluster:
        print(f"  - {task_list[index]}")
    print("---")
```
