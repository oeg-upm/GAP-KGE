```python
import json

with open('../data/model_type.json', 'r') as file:
    data = json.load(file)
```

```python
import os
import json
import requests
import pandas as pd
from urllib.parse import urlparse

os.makedirs("model_type_pdfs", exist_ok=True)
records = []

# 3) Iterate categories and URLs
for entry in data:
    category = entry["category"]
    for url in entry["links"]:
        # derive a safe filename from the URL path
        path = urlparse(url).path
        fname = os.path.basename(path)
        if not fname.lower().endswith(".pdf"):
            # skip non‚Äêpdf or give it a .pdf extension
            fname += ".pdf"
        local_path = os.path.join("model_type_pdfs", fname)

        # 4) Download
        try:
            resp = requests.get(url, stream=True, timeout=30)
            resp.raise_for_status()
            with open(local_path, "wb") as out:
                for chunk in resp.iter_content(1024*32):
                    out.write(chunk)
        except Exception as e:
            print(f"Failed to download {url}: {e}")
            continue

        # 5) Record for CSV
        records.append({
            "category": category,
            "url": url,
            "filename": local_path
        })

# 6) Save to CSV
df = pd.DataFrame(records)
df.to_csv("index.csv", index=False, encoding="utf-8")
print("Done! CSV written to index.csv")
```
